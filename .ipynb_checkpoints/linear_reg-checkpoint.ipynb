## Training

### Standardization
The model can go crazy around its optimization process.

This makes each feature have a mean of 0 and a standard deviation of 1.

**Pros:**
- More efficient optimization
- Ensure the model will converge to a solution

**Standard scaler formula:**
$z = \frac{x-\mu}{\sigma}$

Where:
- $z$ is the standardized value
- $x$ is the original value
- $\mu$ is the mean
- $\sigma$ is the standard deviation

```python
# Initialize the parameters a and b
theta = np.zeros((2, 1))

# Setup X and Y matrices
# Scaling is necessary - standardization method
# All features will have a mean of 0 and a standard deviation of 1.
X = np.array(mileages)
mean_X = np.mean(X)
std_X = np.std(X)
if std_X != 0:
    scaled_X = (X - mean_X) / std_X
    X = np.column_stack((scaled_X, np.ones(X.shape)))
else:
    raise ValueError("Error: Standard deviation of X is zero.")
    
Y = np.array(prices)
mean_Y = np.mean(Y)
std_Y = np.std(Y)
if std_Y != 0:
    scaled_Y = (Y - mean_Y) / std_Y
    Y = Y.reshape(scaled_Y.shape[0], 1)
else:
    raise ValueError("Error: Standard deviation of Y is zero.")
```

**Dataset**: (x, y) avec `m` examples, `n` variables

$X = \begin{bmatrix}x^1 & 1 \\ ... & ... \\ x^m & 1\end{bmatrix}$ $Y = \begin{bmatrix}y^1 \\ ... \\ y^m\end{bmatrix}$ $ \theta = \begin{bmatrix} a \\ b \end{bmatrix}$

**Model:** $ F = X \theta$

**Cost Function:**

$J(\theta) = \frac{1}{2m}\sum(X \theta - Y)^2$

**Gradients:** 

$\frac{\partial{J(\theta}}{\partial{\theta}} = \frac{1}{m}X^T.(X \theta - Y)$

**Gradients Descent:** 

$\theta := \theta - \alpha \frac{\partial{J(\theta)}}{\partial{\theta}}$

- $\alpha$: learning rate

```python
def model(X, theta):
    return X.dot(theta)

def cost_function(X, Y, theta):
    m = len(Y)
    return 1/(2*m) * np.sum((model(X, theta) - Y)**2)

def grad(X, Y, theta):
    m = len(Y)
    return 1 / m * X.T.dot(model(X, theta) - Y)

def gradient_descent(X, Y, theta, learning_rate, n_iterations):
    cost_history = np.empty(n_iterations)
    for i in range(0, n_iterations):
        theta = theta - learning_rate * grad(X, Y, theta)
        cost_history[i] = cost_function(X, Y, theta)
    return theta, cost_history
```

```python
theta_final, cost_history = gradient_descent(X, Y, theta, 0.0001, 60000)

scaled_theta_final = np.zeros_like(theta_final)
# Need to rescale the theta parameters - convert to the original feature scale
# theta_final[0] is the coefficient corresponding to the first feature in the standardized input features (X).
# std_X is the standard deviation of the original feature mileages.
# Dividing theta_final[0] by std_X rescales it to the original scale, effectively undoing the standardization.

if std_X != 0:
    scaled_theta_final[0] = theta_final[0] / std_X
else:
    raise ValueError("Error: Standard deviation of X is zero.")

# theta_final[1] is the coefficient corresponding to the intercept term in the standardized input features (X).
# (mean_X / std_X) is the mean normalization factor used during standardization.
# theta_final[0] corresponds to the coefficient of the first feature, which was also standardized.
# Multiplying (mean_X / std_X) * theta_final[0] adjusts for the mean normalization, essentially bringing it back to the original scale.
# Subtracting this adjustment from theta_final[1] rescales it to the original scale.
if std_X != 0:
    scaled_theta_final[1] = theta_final[1] - (mean_X / std_X) * theta_final[0]
else:
    raise ValueError("Error: Standard deviation of X is zero.")
    
fig, axis = plt.subplots(1, 2, figsize=(16, 4))

color = load_color('mocha', 'peach')

axis[0].set_title("Trained model")
axis[0].set_xlabel('mileage (km)')
axis[0].set_ylabel('price')
axis[0].scatter(mileages, prices)
axis[0].plot(mileages, model(X, theta_final), c=color)
axis[0].set_ylim([0, 8500])

axis[1].set_title('Cost history')
axis[1].set_xlabel('iteration')
axis[1].set_ylabel('cost')
axis[1].plot(cost_history, c=color)

try:
    with open('./model_parameters.txt', 'w') as file:
        file.write(f'{scaled_theta_final[0][0]}\n{scaled_theta_final[1][0]}')
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

## Prediction


Read the model parameters from a text file:

```python
try:
    with open('./model_parameters.txt', 'r') as file:
        lines = file.readlines()
        if len(lines) < 2:
            raise ValueError("File does not contain the expected number of lines.")
        try:
            weight = float(lines[0].strip())
            bias = float(lines[1].strip())
        except ValueError:
            raise ValueError("Non-numeric values found in the file.")
except (FileNotFoundError, ValueError) as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

print(f'bias: {bias}')
print(f'weight: {weight}\n')

# Prompt the user for a mileage
try:
    mileage = float(input('Please input a mileage: '))
    # TODO protect against negative and non numeric
    price = mileage * weight + bias
    print(f'\033[1mEstimated price: {price}\033[0m')
except ValueError:
    print("Invalid input. Please enter a valid number.")
```

## Precision
There are three error metrics that are commonly used for evaluating and reporting the performance of a regression model.

**Metrics:**
- *MSE*: Mean squared error
- *RMSE*: Root Mean squared error
- *MAE*: Mean absolute error
- *R squared*: Coefficient of determination
    - $$R2 = 1 - \frac{SSR}{SST}$$
    - SSR: Sum of squared residuals
    - SST: Sum of squared difference between each data point and the mean.

```python
try:
    with open('./model_parameters.txt', 'r') as file:
        lines = file.readlines()
        if len(lines) < 2:
            raise ValueError("File does not contain the expected number of lines.")
        try:
            weight = float(lines[0].strip())
            bias = float(lines[1].strip())
        except ValueError:
            raise ValueError("Non-numeric values found in the file.")
except (FileNotFoundError, ValueError) as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

Predictions = model(X, theta_final)

plt.scatter(mileages, prices)
plt.plot(mileages, Predictions, c=color)
plt.ylim([0, 8500])
plt.show()

deviations = Y - Predictions

MSE = np.mean(deviations**2)
RMSE = np.sqrt(MSE)
MAE = np.mean(np.abs(deviations))

SSR = np.sum(deviations**2)
SST = np.sum((Y - np.mean(Y))**2)
R2 = 1 - (SSR / SST)

print('\n\033[1mMETRICS:\033[0m')
print('MSE:\t', MSE)
print('RMSE:\t', RMSE, 'km')
print('MAE:\t', MAE)
print('R2:\t', format(R2 * 100, '.4f'), '%')
```
