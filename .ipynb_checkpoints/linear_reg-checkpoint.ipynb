## Read and display data.csv

```python
data_frame = pandas.read_csv('./data.csv')
# print(data_frame)

mileages = data_frame['km']
prices = data_frame['price']

plt.scatter(mileages, prices)
plt.xlabel('kilometers')
plt.ylabel('price')
plt.show()
```

## Utils

## Training

**Dataset**: (x, y) avec `m` examples, `n` variables

$X = \begin{bmatrix}x^1 & 1 \\ ... & ... \\ x^m & 1\end{bmatrix}$ $Y = \begin{bmatrix}y^1 \\ ... \\ y^m\end{bmatrix}$ $ \theta = \begin{bmatrix} a \\ b \end{bmatrix}$

**Model:** $ F = X \theta$

**Cost Function:**

$J(\theta) = \frac{1}{2m}\sum(X \theta - Y)^2$

**Gradients:** 

$\frac{\partial{J(\theta}}{\partial{\theta}} = \frac{1}{m}X^T.(X \theta - Y)$

**Gradients Descent:** 

$\theta := \theta - \alpha \frac{\partial{J(\theta)}}{\partial{\theta}}$

- $\alpha$: learning rate

```python
# Initialize the parameters a and b
theta = np.zeros((2, 1))
# theta = np.random.randn(2, 1)

# Setup matrices
X = np.array(mileages)

# SCALING
mean_X = np.mean(X)
std_X = np.std(X)
scaled_X = (X - mean_X) / std_X

X = np.column_stack((scaled_X, np.ones(X.shape)))

Y = np.array(prices)

mean_Y = np.mean(Y)
std_Y = np.std(Y)
scaled_Y = (Y - mean_Y) / std_Y

Y = Y.reshape(scaled_Y.shape[0], 1)

def model(X, theta):
    return X.dot(theta)

def cost_function(X, Y, theta):
    m = len(Y)
    return 1/(2*m) * np.sum((model(X, theta) - Y)**2)

def grad(X, Y, theta):
    m = len(Y)
    return 1 / m * X.T.dot(model(X, theta) - Y)

def gradient_descent(X, Y, theta, learning_rate, n_iterations):
    cost_history = np.empty(n_iterations)
    for i in range(0, n_iterations):
        theta = theta - learning_rate * grad(X, Y, theta)
        cost_history[i] = cost_function(X, Y, theta)
    return theta, cost_history
```

```python
theta_final, cost_history = gradient_descent(X, Y, theta, 0.001, 40000)

# Need to rescale the theta parameters
scaled_theta_final = np.zeros_like(theta_final)
scaled_theta_final[0] = theta_final[0] / std_X
scaled_theta_final[1] = theta_final[1] - (mean_X / std_X) * theta_final[0]

fig, axis = plt.subplots(1, 2, figsize=(16, 4))

axis[0].set_title("Trained model")
axis[0].set_xlabel('mileage (km)')
axis[0].set_ylabel('price')
axis[0].scatter(mileages, prices)
color = load_color('mocha', 'peach')
axis[0].plot(mileages, model(X, theta_final), c=color)
# TODO: add graph scale
axis[1].set_title('Cost history')
axis[1].set_xlabel('iteration')
axis[1].set_ylabel('cost')
axis[1].plot(cost_history, c=color)

with open('./model_parameters.txt', 'w') as file:
    file.write(f'{scaled_theta_final[0][0]}\n{scaled_theta_final[1][0]}')
```

## Prediction


Read the model parameters from a text file:

```python
with open('./model_parameters.txt', 'r') as file:
    lines = file.readlines()
    weight = float(lines[0].strip())
    bias = float(lines[1].strip())

print(f'bias: {bias} - weight: {weight}')

# Prompt the user for a mileage
try:
    mileage = float(input('Please input a mileage: '))
    # TODO protect against negative and non numeric
    price = mileage * weight + bias
    print(f'Estimated price: {price}')
except ValueError:
    print("Invalid input. Please enter a valid number.")
```
